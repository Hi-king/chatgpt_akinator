```python
import requests
from bs4 import BeautifulSoup
import json
import pathlib
```

```python
datapath = pathlib.Path('../data') / 'imas' / 'source'
```

```python
datapath.mkdir(parents=True, exist_ok=True)
```

```python
response = requests.get('https://ja.wikipedia.org/wiki/%E3%82%A2%E3%82%A4%E3%83%89%E3%83%AB%E3%83%9E%E3%82%B9%E3%82%BF%E3%83%BC_%E3%82%B7%E3%83%B3%E3%83%87%E3%83%AC%E3%83%A9%E3%82%AC%E3%83%BC%E3%83%AB%E3%82%BA%E3%81%AE%E7%99%BB%E5%A0%B4%E4%BA%BA%E7%89%A9')
```

```python
soup = BeautifulSoup(response.content)
```

```python
anchors = soup.find_all(class_='anchor')
```

```python
for anchor in anchors:
    name = anchor.get("id")
    texts = []
    for element in anchor.parent.next_siblings:
        if str(element).find("anchor") >= 0:
            break
        for t in element.text.split("\n"):
            if len(t):
                texts.append(t)
    json.dump(dict(
        name=name,
        texts=texts,
    ), (datapath / f'{name}.json').open('w+'))
```

```python
for p in datapath.glob("*.json"):
    print(json.load(p.open()))
```
